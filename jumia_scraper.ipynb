
### JUMIA SCRAPER

```python
# import the necessary libraries
import pandas as pd
import numpy as np
import requests
import re
import csv
from bs4 import BeautifulSoup

from warnings import filterwarnings
filterwarnings("ignore",category=DeprecationWarning)
filterwarnings("ignore", category=FutureWarning)
```

```python
# Specify the base URL and the number of the pages to scrape
base_url = 'https://www.jumia.com.ng/laptops/'
num_pages = 80

# Create a CSV file to store the scraped data
csv_file = open('laptop_data.csv', 'w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['name', 'actual_price', 'discount_price',' rating'])

# Iterate over each page and scrape the data
for page in range(1, num_pages + 1):
    url = base_url + '?page=' + str(page)
    response = requests.get(url)

    # Create a BeautifulSoup object to parser the html content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the relevant element on the page
    products = soup.find_all('a', class_='core')
    for product in products:
        name_element = product.find('h3')
        price_element = product.find('div', class_='old')
        discount_element = product.find('div', class_='prc')
        rating_element = product.find('div', class_='rev')

        # Extract the data if the elements exist
        name = name_element.text.strip() if name_element else ""
        actual_price = price_element.text.strip() if price_element else ""
        discount_price = discount_element.text.strip() if discount_element else ""
        rating = rating_element.text.strip() if rating_element else ""

        # Write the extracted data to the CSV file
        csv_writer.writerow([name, actual_price, discount_price, rating])

# Close the CSV file
csv_file.close()
```
